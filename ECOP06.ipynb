````python
from typing_extensions import Self
#breed: str, height_low_inches: int, height_high_inches:int, weight_low_lbs:int, weight_high_lbs:int, reps_lower:int, reps_upper:int

class Canino:
  def __init__(self):
    self.breed = ''
    self.height_low_inches = ''
    self.height_high_inches = ''
    self.weight_low_lbs = ''
    self.weight_high_lbs = ''
    self.reps_lower = 0
    self.reps_upper = 0

  def leitura(self,linha):
    v = linha.split(",")
    v = [n.strip() for n in v]
    self.breed = v[0]
    self.height_low_inches = v[1]
    self.height_high_inches = v[2]
    self.weight_low_lbs = v[3]
    self.weight_high_lbs = v[4]
    self.reps_lower = v[5] if len(v)>5 and v[5]!="" else 0
    self.reps_upper = v[6] if len(v)>6 and v[6]!="" else 0

  def __str__(self) -> str:
    return f'{self.breed:30} {self.height_low_inches:4} ' \
    f'{self.height_high_inches:4} {self.weight_low_lbs:4} ' \
    f'{self.weight_high_lbs:4} ' \
    f'{self.reps_lower:4} {self.reps_upper:4}'


class Inteligencia:
  def __init__(self,arquivo):
    self.nomearquivo = arquivo
    self.caninos = []
  def carregar(self):
    with open(self.nomearquivo, 'r', encoding='iso-8859-1') as f:
      cabecalho = f.readline()
      for linha in f:
        c = Canino()
        c.leitura(linha)
        self.caninos.append(c)
  def __str__(self) -> str:
    for c in self.caninos:
      return '\n'.join([str(c) for c in self.caninos])

ic = Inteligencia('Inteligencia_canina.csv')
ic.carregar()
print(ic)
````

Usando a referência em método de tranferência de valor entre duas contas.

````python
class Conta:
  def __init__(self,cpf):
    self.cpf=cpf
    self.corrente=0.0
    self.historico=[]
  def __repr__(self):
    return str(self.__dict__)

  def transfere(self, origem:Conta, valor:float):
    self.corrente += valor
    origem.corrente -= valor
    self.historico.append((origem, valor))
    origem.historico.append((self, -valor))

c1 = Conta('11122233344')
c1.corrente = 500
c2 = Conta('22233344455')
c2.corrente = 1600
c1.transfere(c2, 300)
print('c1 =',c1)
print('c2 =',c2)
````

### Encapsular

Variáveis acessíveis via métodos apenas, com possibilidade de alteração.

````python
from math import sqrt
class Motor:
  def __init__(self,tensao,potencia):
    self._tensao = tensao
    self._potencia = potencia

  def get_tensao(self): # leitura
    return self._tensao

  def get_tensaoPico(self): # leitura alterada
    return self._tensao * sqrt(2)

  def set_tensao(self, tensao): # edição
    if tensao < 0:
      print('Tensão incorreta!')
    else:
      self._tensao = tensao

m = Motor(127,100)
m.set_tensao(-220)
print('Tensão:',m.get_tensao())
print('Tensão Pico:',m.get_tensaoPico())
````

Encapsulando funções de atribuição e entrega:

````python
from math import sqrt
class Motor:
  def __init__(self,tensao,potencia):
    self._tensao = tensao
    self._potencia = potencia

  @property
  def tensao(self): # leitura
    return self._tensao

  def get_tensaoPico(self): # leitura alterada
    return self._tensao * sqrt(2)

  @tensao.setter
  def tensao(self, tensao): # edição
    if tensao < 0:
      print('Tensão incorreta!')
    else:
      self._tensao = tensao

m = Motor(127,100)
m.tensao = -220
print('Tensão:',m.tensao)
print('Tensão Pico:',m.get_tensaoPico())
````

### SLOTS

````python
class Endereco:
  __slots__ = ['continente','pais','estado','cidade',
               'bairro','logradouro','numero',
               'complemento','cep','__dict__']

  def __init__(self, continente, pais, estado, cidade,
               bairro, logradouro, numero, complemento, cep):

    self.continente=continente
    self.pais=pais
    self.estado=estado
    self.cidade=cidade
    self.bairro=bairro
    self.logradouro=logradouro
    self.numero=numero
    self.complemento=complemento
    self.cep=cep

end = Endereco('america','brasil','minas gerais','itajubá','pinheirinho','avenida bps',1303,'','37500903')
end.bairro

end.__dict__
````

**prós**

\_\_slots\_\_ pode ser definitivamente útil quando você tem uma pressão sobre o uso da memória.

É extremamente fácil adicionar ou remover com apenas uma linha de código.

A possibilidade de ter \_\_dict\_\_ como atributo em \_\_slots\_\_ dá aos desenvolvedores mais flexibilidade para gerenciar atributos enquanto cuidam do desempenho.

**contras**

Você precisa ser claro sobre o que está fazendo e o que deseja alcançar com o \_\_slots\_\_, especialmente ao herdar uma classe com ele.

A ordem de herança, os nomes dos atributos podem fazer uma grande diferença no desempenho.

Você não pode herdar um tipo interno como int, bytes, tuple não vazio \_\_slots\_\_.

Além disso, você não pode atribuir um valor padrão aos atributos em \_\_slots\_\_. Isso ocorre porque esses atributos devem ser descritores. Em vez disso, você pode atribuir o valor padrão em \_\_init\_\_().

## Herança de Classes

[Documentação](https://www.geeksforgeeks.org/inheritance-in-python/)

[Oficial](https://docs.python.org/pt-br/3.8/tutorial/classes.html)

Herança é a capacidade de uma classe derivar ou herdar as propriedades de outra classe.

Os benefícios da herança são:
- Representa bem os relacionamentos do mundo real.
- Ele fornece a reutilização de um código. Não temos que escrever o mesmo código repetidamente. Além disso, permite adicionar mais recursos a uma classe sem modificá-la.
- É de natureza transitiva, o que significa que, se a classe B herdar de outra classe A, todas as subclasses de B herdarão automaticamente da classe A.

````python
import abc

class Conta(abc.ABC):
    def __init__(self, saldo):
        self.saldo = saldo

    @abc.abstractmethod
    def atualiza(self, taxa):
        pass

    @abc.abstractmethod
    def deposita(self, valor):
        pass

class ContaCorrente(Conta):
    def atualiza(self, taxa):
        self.saldo += self.saldo * (2 * taxa / 100)

    def deposita(self, valor):
        valor -= 0.10
        self.saldo += valor

class ContaPoupanca(Conta):
    def atualiza(self, taxa):
        self.saldo += self.saldo * (3 * taxa / 100)

    def deposita(self, valor):
        self.saldo += valor

contas = [ContaCorrente(10_000), ContaPoupanca(10_000)]

for c in contas:
    c.atualiza(5)  # O cara recebe 5% de comissão por mês
    c.deposita(100)  # O cara bota 100 por mês

[(c.__class__.__name__, c.saldo) for c in contas]
````

Composição é um conceito que modela uma relação.

As classes que contêm objetos de outras classes geralmente são chamadas de compostos, onde as classes usadas para criar tipos mais complexos são chamadas de componentes.

````python
class Component:
  def __init__(self):
    print('Component class object created...')

  def m1(self):
    print('Component class m1() method executed...')

class Composite:
	def __init__(self):
		self.obj1 = Component()
		print('Composite class object also created...')

	def m2(self):
		print('Composite class m2() method executed...')
		self.obj1.m1()

obj2 = Composite()
obj2.m2()
````

**Proposta**: construa um conjunto de classes para armazenar os dados de previsão para um conjunto de cidades fornecido.

````python
import requests
import urllib.parse
from xml.etree import ElementTree

def ExtractContent(content):
  arvore = ElementTree.fromstring(content)
  itens = []
  for ramo in arvore:
    item = {}
    for folha in ramo:
      item[folha.tag] = folha.text
    if len(item.keys()) > 0:
      itens.append(item)
  return itens

# classe de previsao
class Previsao:
  def __init__(self, cidade_id):
    self.cidade_id = cidade_id
    prev_url = f'http://servicos.cptec.inpe.br/XML/cidade/{cidade_id}/previsao.xml'
    response = requests.get(prev_url)
    self.dias = ExtractContent(response.content)
  def __repr__(self):
    saida = ''
    for dia in self.dias:
      saida+= '{} [máxima: {} mínima: {}] {} \n'.format(
          dia['dia'],dia['maxima'],dia['minima'],dia['tempo'],
      )
    return saida

# criando a classe que busca a cidade
class Cidade:
  def __init__(self, cidade_nome:str):
    self.cidade_nome = cidade_nome
    self.cidade_inpe = []
    self.previsao = None
  def carregar(self):
    self.city_url = 'http://servicos.cptec.inpe.br/XML/listaCidades?city=' + urllib.parse.quote(self.cidade_nome)
    response = requests.get(self.city_url)
    # pegando a 1a cidade
    self.cidades_inpe = ExtractContent(response.content)
    if len(self.cidades_inpe) > 0:
      self.previsao = Previsao(self.cidades_inpe[0]['id'])
    return self
  def __repr__(self):
    if len(self.cidades_inpe) > 0:
      saida = '{}-{}'.format(self.cidades_inpe[0]['nome'],self.cidades_inpe[0]['uf'])
      saida += '\n'
      saida +=  self.previsao.__repr__()
      return saida
    else:
      return 'Cidade não encontrada'

cidadesVetor = ['Aparecida', 'Guaratingueta', 'taubate', 'potim']

for i in cidadesVetor:
  ci = Cidade(i)
  ci.carregar()
  print(ci)
````
